```python

print(df['Adult Mortality'].describe())

plt.figure(figsize=(8,5))
df['Adult Mortality'].dropna().plot(kind='hist', bins=20)
plt.title('Distribution of Adult Mortality')
plt.xlabel('Adult Mortality')
plt.ylabel('Number of Records')
plt.show()
```

```python
df.columns = df.columns.str.strip()
print(df.columns.tolist())
```

```python
correlation = df['Adult Mortality'].corr(df['Life expectancy'])
print("Correlation between Adult Mortality and Life expectancy:", correlation)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Adult Mortality'], df['Life expectancy'])
plt.title('Adult Mortality vs Life Expectancy')
plt.xlabel('Adult Mortality')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
df.columns = df.columns.str.strip()

temp = df[['Alcohol', 'Life expectancy']].dropna()

plt.figure(figsize=(8,5))
plt.scatter(temp['Alcohol'], temp['Life expectancy'])
plt.title('Alcohol vs Life Expectancy')
plt.xlabel('Alcohol Consumption')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
df.columns = df.columns.str.strip()

temp = df[['Alcohol', 'Life expectancy']].dropna()
x = temp['Alcohol']
y = temp['Life expectancy']

plt.figure(figsize=(8,5))
plt.scatter(x, y)

m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)

plt.title('Alcohol vs Life Expectancy')
plt.xlabel('Alcohol Consumption')
plt.ylabel('Life Expectancy')
plt.show()
```
